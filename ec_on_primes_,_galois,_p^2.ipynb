```python
# Define the parameters of the elliptic curve
a = 0
b = 6
p = 11  # the field size

# Define the base point G
Gx = 2
Gy = 4

# Define a function to add two points on the curve
def point_addition(P, Q):
    if P is None:
        return Q
    if Q is None:
        return P
    if P[0] == Q[0] and P[1] != Q[1]:
        return None

    if P == Q:
        lam = ((3 * P[0] * P[0] + a) * pow(2 * P[1], p - 2, p)) % p
    else:
        lam = ((Q[1] - P[1]) * pow(Q[0] - P[0], p - 2, p)) % p

    Rx = (lam * lam - P[0] - Q[0]) % p
    Ry = (lam * (P[0] - Rx) - P[1]) % p

    return (Rx, Ry)

# Define a function to double a point on the curve
def point_doubling(P):
    if P is None:
        return None
    lam = ((3 * P[0] * P[0] + a) * pow(2 * P[1], p - 2, p)) % p
    Rx = (lam * lam - 2 * P[0]) % p
    Ry = (lam * (P[0] - Rx) - P[1]) % p
    return (Rx, Ry)

# Define a function to compute the order of a point on the curve
def point_order(P):
    if P is None:
        return 1
    Q = P
    order = 1
    while(Q != None):
        Q = point_addition(P, Q)
        order = order + 1
        if(order > p+1):
            return None
    return order

# Define a function to find all points on the curve with a given order
def find_points_with_order(order):
    points = []
    for x in range(p):
        y2 = (x**3 + a*x + b) % p
        if pow(y2, (p-1)//2, p) == 1:
            y = pow(y2, (p+1)//4, p)
            P = (x, y)
            Q = P
            while Q is not None and point_order(Q) != order:
                Q = point_addition(Q, P)
            if point_order(Q) == order:
                points.append(Q)
    return points

print(find_points_with_order(12))
```

```python
# define the irreducible polynomial and field size
irr_poly = 0b10011  # x^4 + x + 1 and 0b is a prefix used in Python to indicate a binary literal
field_size = 2 ** 4  # GF(2^4)

# generate the elements of the Galois field
elements = []
for i in range(field_size):
    element = i
    while element >= field_size:
        element ^= irr_poly       # ''^'' XOR operation
        element &= field_size - 1 # AND operation
    elements.append(element)      # append() is a list method that adds an element to the end of the list.

# print the elements of the Galois field
for element in elements:
    print(element)
```
