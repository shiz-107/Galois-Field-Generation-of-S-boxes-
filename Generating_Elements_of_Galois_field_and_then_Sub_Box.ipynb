```python
def binary_gcd(a, b):
    if a == 0:
        return b
    if b == 0:
        return a
    k = 0
    while ((a | b) & 1) == 0:
        a >>= 1
        b >>= 1
        k += 1

    while (a & 1) == 0:
        a >>= 1

    while b != 0:
        while (b & 1) == 0:
            b >>= 1
        if a > b:
            a, b = b, a
        b = b - a

    return a << k

def inverse_in_gf(element):
    modulus = 0b100011101  # GF(2^8) modulus
    element = element & 0xFF
    if element == 1:
        return 1
    inverse = binary_gcd(modulus, element)
    return inverse

def A(z):

    a =54
    b =90
    c = 1
    d = 97
    field_size = 256
    irr_poly = 0b100011101

    num = a * z + b
    while num >= field_size:
        num ^= irr_poly
        num &= field_size - 1

    den = c * z + d
    while den >= field_size:
        den ^= irr_poly
        den &= field_size - 1
    inverse_den = inverse_in_gf(den)
    result = (num * inverse_den) % field_size
    while result >= field_size:
        result ^= irr_poly
        result &= field_size - 1
    return result

galois_elements = range(256)
results = [A(z) for z in galois_elements]
print(results)

```
